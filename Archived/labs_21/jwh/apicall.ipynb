```python
from google.cloud import vision
```

```python
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "ProjectKey.json"
```

```python
client = vision.ImageAnnotatorClient()
```

```python
import io

path = './recipeimages/download.jpg'
with io.open(path, 'rb') as image_file:
        content = image_file.read()
```

```python
image = vision.types.Image(content=content)
```

```python
response = client.document_text_detection(image=image)
```

```python
texts = response.text_annotations
result = response.full_text_annotation.text
```

```python
response = client.document_text_detection(image=image)
```

```python
print(f'{len(response.full_text_annotation.pages[0].blocks)} blocks') 
for block in response.full_text_annotation.pages[0].blocks:
    print('----------------------')
    blocktext = []
    for paragraph in block.paragraphs:
        for word in paragraph.words:
            wordtext = ''
            for symbol in word.symbols:
                wordtext += symbol.text
            blocktext.append(wordtext)
    print(' '.join(blocktext), '\n')    
```
