## Get the URLs for the recipes on allrecipes.com

### Import BeautifulSoup and requests

```python
from bs4 import BeautifulSoup
import requests
```

```python
# I'm trying to get all of the recipes off of allrecipes.com.
# This is the sitemap of the recipe categories.
main_url = 'https://www.allrecipes.com/recipes/'
```

```python
# Make a call to get the html of main_url
response = requests.get(main_url)
soup = BeautifulSoup(response.text, "lxml")
```

## Get the links for each of the main categories

```python
# They are contained in the class "col-container"
soup.find('div', class_='col-container')
```

```python
# The recipes are in three columns divided into 'all-categories-col'. 
# This is just another way to get all of the parent links.
categories = soup.find_all('div', class_='all-categories-col')
categories
```

```python
# Parent links for different categories of recipes
parent_cat_urls = []

for ul in soup.find_all('div', class_='all-categories-col'):
  for li in ul.find_all('li'):
    a = li.find('a')
    print(a['href'], a.get_text())
    parent_cat_urls.append(a['href'])
```

### The pages are infinite scrolling so I'm missing recipes and want more than the 'first' page shows automatically.

```python
# Make a set to store recipe urls. A set will make it so a given recipe is only
# saved once in case it's stored in more than one category.
recipe_urls_set = set()
```

```python
# Set up a loop to check for valid webpages to move through infinite scrolling
i = 1  # Start with page 1

while True:
  # I got this url in the Request Headers section of the network dev tools
  # in Google Chrome after selecting just XHR.
  url = f'https://www.allrecipes.com/recipes/22951/breakfast-and-brunch/pancakes/buttermilk-pancakes/?page={i}'
  r = requests.get(url)
  #print(r)
  if r.status_code == 404:  # Check if 404 error when trying to load the ith page.
    break  # Break out of loop if ith page doesn't exist.
  else:  # Continue if it does.
    soup = BeautifulSoup(r.text, "lxml")
    #print(i)

    i += 1  # Increment to go to next page.


    for article in soup.find_all('div', class_='fixed-recipe-card__info'):
      link = article.h3.a['href']
      recipe_urls_set.add(link)  # Add the link to set if it's new.
```

## Time to try getting all of the recipes using the above logic and loops.

```python
# I want to be kind to the allrecipes.com servers so include pauses in scraping.
import time
t = 0.5
```

```python
# All of the parent categories on https://www.allrecipes.com/recipes/
len(parent_cat_urls)
```

```python
# Set of child categories.
child_cat_urls_set = set()
```

```python
for i in range(len(parent_cat_urls)):
  response = requests.get(parent_cat_urls[i])
  soup = BeautifulSoup(response.text, "lxml")
  print(f"Getting children links for {parent_cat_urls[i]}. {i+1} of 92.")

  for ul in soup.find_all('div', class_='grid slider'):
    for li in ul.find_all('li'):
      a = li.find('a')
  #    print(a['href'], a.get_text())
      child_cat_urls_set.add(a['href'])
  
  time.sleep(t) # Pause before continuing
```

```python
len(child_cat_urls_set)
```

```python
# Grandchild categories set.
grandchild_cat_urls_set = set()
```

```python
for i, child_cat in enumerate(child_cat_urls_set):
  response = requests.get(child_cat)
  soup = BeautifulSoup(response.text, "lxml")
  print(f"Getting grand-children links for {child_cat}. {i+1} of 974.")

  for ul in soup.find_all('div', class_='grid slider'):
    for li in ul.find_all('li'):
      a = li.find('a')
      #print(a['href'], a.get_text())
      grandchild_cat_urls_set.add(a['href'])

  time.sleep(t) # Pause before continuing
```

```python
# Total grandchild categories where the recipes are actually kept.
len(grandchild_cat_urls_set)
```

## The category urls are stored. Now get the recipe urls.

```python
# Set for all of the recipe urls and to avoid duplicates
recipe_set = set()
```

# This next cell is going to take a LONG time.

```python
for j, grandchild_url in enumerate(grandchild_cat_urls_set):
  print(f'Getting recipe links from {grandchild_url}. {j+1} of 1708')
  i = 1

  while True:
    url = f'{grandchild_url}?page={i}'
    r = requests.get(url)
    #print(r)
    if r.status_code == 404:
      break
    else:
      soup = BeautifulSoup(r.text, "lxml")
      #print(i)

      i += 1

      for article in soup.find_all('div', class_='fixed-recipe-card__info'):
        link = article.h3.a['href']
        recipe_set.add(link)

      time.sleep(0.5)
```

```python
# How many recipes total?
len(recipe_set)
```

```python
recipe_set
```

```python
import csv
```

```python
# Save the recipe urls to a csv for later use.
with open('allrecipes_urls.csv', 'w', newline='') as f:
  thewriter = csv.writer(f)

  for recipe in recipe_set:
    thewriter.writerow([recipe])
```
